```python
import re
import pandas as pd
from langdetect import detect, LangDetectException

files = ['eng', 'amh', 'arb', 'deu', 'hau', 'ita', 'spa', 'urd', 'zho']

def clean_text(t):
    t = str(t).lower()
    t = re.sub(r"http\S+|www\S+|https\S+", "", t)
    t = re.sub(r"@\w+", "@user", t)
    t = re.sub(r"#\w+", "", t)
    t = re.sub(r"&\w+;", "", t)
    t = re.sub(r"[^\w\s]", "", t)
    t = re.sub(r"\s+", " ", t).strip()
    return t

def safe_detect(text):
    try:
        if isinstance(text, str) and len(text.strip()) > 10:
            return detect(text)
        else:
            return "unknown"
    except LangDetectException:
        return "unknown"

for lang in files:
    file = f"/content/raw/{lang}.csv"
    df = pd.read_csv(file)

    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('/', '_')
    df = df.dropna(subset=['text'])

    df['clean_text'] = df['text'].apply(clean_text)
    df['lang_detected'] = df['clean_text'].apply(safe_detect)

    df.to_csv(f"/content/processed/{lang}_clean.csv", index=False)
    print(f"Saved cleaned + language-detected file: {lang}_clean.csv")
```

```python
import os
import pandas as pd
import spacy
import matplotlib.pyplot as plt
import seaborn as sns

os.makedirs("/content/pdf", exist_ok=True)

# Load spaCy model (English; you can swap with 'xx_ent_wiki_sm' for multilingual)
nlp = spacy.load("en_core_web_sm")

files = ['eng', 'amh', 'arb', 'deu', 'hau', 'ita', 'spa', 'urd', 'zho']

for lang in files:
    file = f"/content/processed/{lang}_clean.csv"
    df = pd.read_csv(file)

    pos_counts = {}
    ner_counts = {}

    # Process a sample for speed
    sample = df['clean_text'].dropna().astype(str).head(300)

    for doc in nlp.pipe(sample, disable=["parser"]):
        for token in doc:
            pos_counts[token.pos_] = pos_counts.get(token.pos_, 0) + 1
        for ent in doc.ents:
            ner_counts[ent.label_] = ner_counts.get(ent.label_, 0) + 1

    # POS Plot
    plt.figure(figsize=(8, 5))
    sns.barplot(x=list(pos_counts.keys()), y=list(pos_counts.values()), palette="viridis")
    plt.title(f"POS Tag Counts ({lang})")
    plt.tight_layout()
    plt.savefig(f"/content/pdf/POS_counts_{lang}.pdf")
    plt.show()

    # NER Plot
    plt.figure(figsize=(8, 5))
    sns.barplot(x=list(ner_counts.keys()), y=list(ner_counts.values()), palette="magma")
    plt.title(f"Named Entity Counts ({lang})")
    plt.tight_layout()
    plt.savefig(f"/content/pdf/NER_counts_{lang}.pdf")
    plt.show()

    print(f"Saved POS & NER plots for {lang}")
```
